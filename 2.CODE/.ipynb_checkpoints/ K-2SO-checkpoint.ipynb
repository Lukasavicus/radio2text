```python
import os
from datetime import datetime
import pandas as pd
from utils import combine_audio_files
```

### 2. Loading Data

```python
# _datetime = '2024-12-11-11-00'  # OK
# _datetime = '2024-12-11-12-00'  # OK
# _datetime = '2024-12-11-13-30'  # OK
# _datetime = '2024-12-12-08-00'  # OK
_datetime = '2024-12-13-00-00'    # NOK
# _datetime = '.'
audios_path = f'1.bronze/{_datetime}/audio_chunks'
transcriptions_path = f'1.bronze/{_datetime}/transcription_chunks'
files_data = []

transcriptions = [filename for filename in os.listdir(transcriptions_path) if filename.endswith('.txt')]
transcriptions.sort()
audios = [filename for filename in os.listdir(audios_path) if filename.endswith('.mp3')]
audios.sort()

try:
    assert len(transcriptions) == len(audios), f"len(transcriptions)={len(transcriptions)} != len(audios)={len(audios)}"
except AssertionError:
     print(f"Warning - Number of transcriptions and audios do not match.\n\ttranscriptions:{len(transcriptions)}\n\taudios:{len(audios)}")


for filename in transcriptions:
        file_path = os.path.join(transcriptions_path, filename)
        audio_file_path = os.path.join(audios_path, filename.split('.')[0] + '.mp3')
        with open(file_path, 'r') as file:
            content = file.read()
        timestamp = os.path.getmtime(file_path)
        timestamp = datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')
        files_data.append({
            'name_of_file': filename,
            'content': content,
            'audio_file': audio_file_path,
            'timestamp_of_file': timestamp
        })

print(files_data)
```

```python
df = pd.DataFrame(files_data)
```

```python
def look_for_words(words, excerpt):
    return any(word in excerpt for word in words)

words = ['caçada', 'palavra', 'chave']

f = df['content'].apply(lambda x: look_for_words(words, x))
idxs = list(df[f].index)
df[f]
```

```python
def group_indices(indices, lookup_window, extreme_window=2, min_indice=0, max_indice=None, min_group_size=None, max_group_size=None):
    if not indices:
        return []
    
    if extreme_window >= lookup_window:
        raise ValueError("extreme_window must be less than lookup_window") # Or should be a warning?

    if max_indice is None: max_indice = indices[-1] + extreme_window + 1
    grouped_indices = []
    current_group = [indices[0]]

    for i in range(1, len(indices)):
        if indices[i] - indices[i - 1] <= lookup_window:
            current_group.append(indices[i])
        else:
            current_group = [i for i in range(current_group[0] - extreme_window, current_group[-1] + extreme_window + 1) if min_indice <= i < max_indice]
            grouped_indices.append(current_group)
            current_group = [indices[i]]
    
    current_group = [i for i in range(current_group[0] - extreme_window, current_group[-1] + extreme_window + 1) if min_indice <= i < max_indice]
    grouped_indices.append(current_group)
    return grouped_indices

# Exemplo de uso
# O que fazer nesse caso?
# indices = [1, 2, 3, 10, 11, 20, 21, 22, 30]
# print(group_indices(indices, 2, 4))

indices = [1, 2, 3, 10, 11, 20, 21, 22, 30]
print(group_indices(indices, 7, 1))
```

```python
gi = group_indices(idxs, 4, 2)
dest_folder = f'./2.silver/{_datetime}/audio_chunks'
os.makedirs(dest_folder, exist_ok=True) 
for group in gi:
    audios = list(df.iloc[group]['audio_file'])
    filename = f"f_{group[0]}_{group[-1]}.mp3"
    combine_audio_files(audios, f'{dest_folder}/{filename}')
```
